### Find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

### Create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

### Checking if deaths have the same correlations

```python
corona_dataset_csv1 = pd.read_csv(r"C:\Users\wojci\OneDrive\Pulpit\Portfolio\Covid19_Data_Analysis\covid19_deaths_dataset.csv")
corona_dataset_csv1.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
corona_dataset_aggregated1 = corona_dataset_csv1.groupby("Country/Region").sum()
corona_dataset_aggregated1.head() 
```

```python
countries1 = list(corona_dataset_aggregated1.index)
death_rate = []
for c in countries1:
    death_rate.append(corona_dataset_aggregated1.loc[c].diff().max())
corona_dataset_aggregated1["death_rate"] = death_rate
```

```python
corona_dataset_aggregated1.head()
```

```python
corona_data1 = pd.DataFrame(corona_dataset_aggregated1["death_rate"])
corona_data1.head()
```

```python
data1 = corona_data1.join(happiness_report_csv,how="inner")
data1.head()
```

```python
data1.corr()
```

```python
x = data1["GDP per capita"]
y = data1["death_rate"]
sns.regplot(x=x, y=np.log(y))
```

```python
x = data1["GDP per capita"]
y = data1["death_rate"]
sns.scatterplot(x=x, y=np.log(y))

z = np.polyfit(x, np.log(y), 2)
p = np.poly1d(z)

plt.plot(x, p(x))
```
