## Python Project for movies dataset correlations

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:/Users/wojci/OneDrive/Pulpit/Portfolio/movies.csv')
df.head()
```

```python
# We have to check missing values
df.head()
df.isnull().sum()
```

```python
# I will delete missing values but we can also fill them, by random forest, median there are many ways

df = df.dropna()
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
# I don't like datatypes of budget, votes and gross, it does not look good in dataframe

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Checking movies with the biggest gross
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Dropping duplicates
df.drop_duplicates()
```

```python
# I assume that budged and gross earnings have high correlation so i will try to check it
plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# Let's check correlation between variables. IMPORTANT NOTE!! -- checking corr will only work on numerical columns, it will not consider columns like company, writer, name etc.

df.corr(method ="pearson") # kendall, spearman
```

```python
#As I assumed, budget and gross earnings have high correlation

matrix_corr = df.corr(method ='pearson')

sns.heatmap(matrix_corr, annot=True)

plt.title("Correlation Matric for numeric variables")
plt.show()
```

```python
# We can also numerize string/text columns if we want to use them in correlation matrix
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
numerized_matrix_corr = df_numerized.corr(method ='pearson')

sns.heatmap(numerized_matrix_corr, annot=True)

plt.title("Correlation Matric for numeric variables")
plt.show()
```

```python
# It's hard to see anything so let's unstack and sort it
df_numerized.corr(method ='pearson')
```

```python
# Also votes have high correlation to gross
corr_pairs = numerized_matrix_corr.unstack()

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,                 
                       ):print(corr_pairs[corr_pairs>0.5].sort_values())
```
